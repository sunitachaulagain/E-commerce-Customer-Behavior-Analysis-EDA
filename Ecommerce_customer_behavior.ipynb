### EDA E-commerce Customer Behavior Analysis

### Problem Statement

This project aims to analyze customer behavior in an e-commerce platform by examining factors such as demographic attributes (age, gender, location), engagement metrics (login frequency, session duration, pages per session), and purchasing behavior (cart abandonment, wishlist activity, lifetime value), in order to identify patterns related to customer retention, high-value users, and churn.

### Data Collection

* Dataset Source: https://www.kaggle.com/datasets/dhairyajeetsingh/ecommerce-customer-behavior-dataset

* The data consists of 25 columns and 50,000 rows.

3. DataSet Information

* Age : Age of the customer
* Gender : Gender of the customer (Male/Female)
* Country : Country where the customer is located
* City : City of the customer
* Membership_Years : Number of years the customer has been using the platform
* Login_Frequency : Number of times the customer logs into the platform
* Session_Duration_Avg : Average time spent per session
* Pages_Per_Session : Average number of pages viewed per session
* Cart_Abandonment_Rate : Percentage of times a customer adds items to cart but does not complete purchase
* Wishlist_Items : Number of items added to wishlist
* Email_Open_Rate : Percentage of marketing emails opened by the customer
* Customer_Service_Calls : Number of times the customer contacted support
* Product_Reviews_Written : Number of reviews written by the customer
* Social_Media_Engagement_Score : Level of interaction on social media platforms
* Mobile_App_Usage : Usage level of the mobile application
* Payment_Method_Diversity : Number of different payment methods used
* Lifetime_Value : Total revenue generated from the customer
* Credit_Balance : Available credit balance of the customer
* Churned : Whether the customer has stopped using the platform (0 = No, 1 = Yes)
* Signup_Quarter : Quarter in which the customer signed up (Q1, Q2, Q3, Q4)

### 1. Handle Missing Values

*  checking the missing value in dataset


```python

missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({'missing_value' : missing_values,
                           'percentage'  : missing_percentage})

print(missing_df)
```

* Handle missing values

1. Columns to Fill with Median
* Continuous numeric columns where missing = unknown value:
    * Age → median
    * Session_Duration_Avg → median
    * Pages_Per_Session → median
    * Days_Since_Last_Purchase → median

```python
num_median_cols = ['Age', 'Session_Duration_Avg', 'Pages_Per_Session', 'Days_Since_Last_Purchase']

for col in num_median_cols:
    df[col] = df[col].fillna(df[col].median())
```

2. Columns to fill with 0

* Missing likely means “didn’t do it / no activity”:

    * Wishlist_Items
    * Discount_Usage_Rate
    * Returns_Rate
    * Email_Open_Rate
    * Customer_Service_Calls
    * Product_Reviews_Written
    * Mobile_App_Usage
    * Social_Media_Engagement_Score
    * Credit_Balance



```python
zero_fill_cols = ['Wishlist_Items','Discount_Usage_Rate','Returns_Rate','Email_Open_Rate','Customer_Service_Calls','Product_Reviews_Written','Mobile_App_Usage','Social_Media_Engagement_Score','Credit_Balance']

for col in zero_fill_cols:
    df[col] = df[col].fillna(0)
```

3. Categorical / Small Missing Columns

* Payment_Method_Diversity → fill with median or mode (usually small % missing, 5%)

```python
df['Payment_Method_Diversity'] = df['Payment_Method_Diversity'].fillna(df['Payment_Method_Diversity'].median())
```

```python
# checking missing values again

df.isna().sum()
```

```python
df.notna().sum()
```

### Insight or Observation

* Out of 25 columns, 14 columns have missing values.
* Most missing values are less than 10% of the total data(50,000 rows), so we can safely fill them without dropping columns.
* Columns like Social_Media_Engagement_Score (12%) and Credit_balance (11%) have slightly higher missing values, but still managable.
* Missing values can be handled by:
    * FIlling continuous numeric columns with median
    * Filling activity/count columns with 0
    * Filling categorical columns with mode
